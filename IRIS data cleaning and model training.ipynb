```python
# Import the neccessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split    
```

```python
#load the dataset
#view the first 5 rows of the dataset
iris = pd.read_csv('iris.csv')
print(iris.head())
```

```python
#check data properties
iris.info()
```

```python
iris.describe()
```

```python
iris.shape
```

```python
iris.isnull().sum()
```

```python
iris['species'].unique()
```

```python
#encode the target variable
iris['species'] = iris['species'].map({'Iris-setosa':0,'Iris-versicolor':1, 'Iris-virginica':2})
iris.head()
```

### Split Data

X = Feature (input varaible)

y = Target (what we want to predict)

```python
X = iris.drop('species', axis=1)
y = iris['species']
```

```python
X
```

```python
y
```

```python
# train_test_split is used to split the dataset into training and testing sets. It takes the features (X) and target variable (y) as input, along with the test size and random state for reproducibility. The function returns four sets: X_train, y_train, X_test, and y_test.    
# - X_train: The features for the training set.
# - y_train: The target variable for the training set. 
# - X_test: The features for the testing set.
# - y_test: The target variable for the testing set.
# The test_size parameter specifies the proportion of the dataset to include in the test split (in this case, 33% for testing and 67% for training). The random_state parameter ensures that the split is reproducible, meaning that the same random state will yield the same split each time the code is run.


(X_train, X_test, y_train, y_test) = train_test_split(X,y,test_size=0.33, random_state=42)
X_train
```

```python
y_train
```

### Logistic Regression Algorithm

```python
# Intstantiate/launche the LR model
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
```

```python
# train the model
logreg.fit(X_train, y_train)
```

```python
# prediction

y_pred_logreg = logreg.predict(X_test)
```

```python
y_pred_logreg
```

```python
# evaluation
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("accuracy_score_logreg: ", accuracy_score(y_pred_logreg, y_test))
print("\nconfusion_matrix_logreg")
print(confusion_matrix(y_pred_logreg, y_test))
print("\nclassification_report_logreg")
print(classification_report(y_pred_logreg, y_test))
```

### Random Forest Classifier Algorithm

```python
# Intstantiate/launche the LR model
from sklearn.ensemble import RandomForestClassifier 
rfc = RandomForestClassifier()
```

```python
#train the model
rfc.fit(X_train, y_train)
```

```python
# predictions
y_pred_rfc = rfc.predict(X_test)
```

```python
y_pred_rfc
```

```python
# evaluation

print("accuracy_score_rfc: ", accuracy_score(y_pred_rfc, y_test))
print("\nconfusion_matrix_rfc")
print(confusion_matrix(y_pred_rfc, y_test))
print("\nclassification_report_rfc")
print(classification_report(y_pred_rfc, y_test))
```

### Gradient Boosting Classifier Algrithm

```python
# Intstantiate/launche the gbc model
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier()
```

```python
#train the model
gbc.fit(X_train, y_train)
```

```python
#prediction
y_pred_gbc = gbc.predict(X_test)
```

```python
y_pred_gbc
```

```python
#evaluation

print("accuracy_score_gbc: ", accuracy_score(y_pred_gbc, y_test))
print("\nconfusion_matrix_gbc")
print(confusion_matrix(y_pred_gbc, y_test))
print("\nclassification_report_gbc")
print(classification_report(y_pred_gbc, y_test))
```

### K Nearest Neighbors (KNN) Algorithm

```python
# Intstantiate/launche the knn model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
#train the model
knn.fit(X_train, y_train)
```

```python
#prediction
y_pred_knn = knn.predict(X_test)
```
