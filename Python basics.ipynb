### Functions
- you can choose to not have any argument
- you can choose to have one or more parameters/arguments
- when function is called, you can choose to pass arguments as positional or named

```python
## positional argument vs name argument explained

def min_three_numbers (first_number, second_number, third_number):
    if first_number < second_number:
        min_num = first_number
        if first_number < second_number:
            min_num = first_number
        else:
            min_num = third_number
    elif second_number < first_number:
        min_num = second_number
        if second_number < third_number:
            min_num = second_number
        else:
            min_num = third_number
    return min_num
```

# Mini Project: Simple ATM System

Build a Simple ATM System

Requirements:
- Ask for balance

- Ask for withdrawal amount

- Check if enough money

- Use a function

```python
def atm_withdrawal(current_balance, withdrawal_amount):

    if withdrawal_amount <= current_balance:
        output = current_balance - withdrawal_amount
    else:
        output = "Insufficient funds."
    print(output)
```

```python
current_balance = eval(input("Enter your current balance:"))
withdrawal_amount = eval(input("Enter your withdrawal amount:"))

atm_withdrawal(current_balance, withdrawal_amount)
```

3. Create a guessing game using a while loop

```python
# 1. Start with your introduction

print("Welcome! This is a guessing game.")
print("I'm thinking of a number between 1 and 10...")

# 2. Set up your variables

secret = 7
guess = 0 # Starting at 0 so the loop begins

# 3. Start the loop

while guess != secret:
    # Ask for the guess INSIDE the loop
    guess = eval(input("Enter your guess: "))

    # Check if it's too high or too low
    if guess < secret:
        print("Too low, try again!")
    elif guess > secret:
        print("Too high, try again!")

# 4. Final message 
# This is outside the loop - it only runs when guess == secret

print("Success! You guessed right.")
```
